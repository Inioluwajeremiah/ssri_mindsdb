```python
ki = pd.read_csv('serotonin_clean_needed_dataset.csv')
ki = ki[['standard_value']]
ki.shape
```

```python
# using minmaxscaler to normalize

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
print("scaler fit", scaler.fit(ki))
print("scaler data_max", scaler.data_max_)
new_y = pd.DataFrame(scaler.transform(ki))
new_y.to_csv("serotonin_ki_minmax_scale.csv", index=False)
new_y.describe()
```
